```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart_disease_uci.csv')
df.head()
```

```python
# Menampilkan 5 baris pertama
print("5 Baris Pertama:")
print(df.head())
```

```python
# Menampilkan 5 baris terakhir
print("\n5 Baris Terakhir:")
print(df.tail())
```

```python
# Melihat informasi tentang dataset
print("\nInformasi Dataset:")
print(df.info())
```

```python
# Cek nilai missing
print("\nJumlah Missing Values:")
print(df.isnull().sum())
```

```python
# Cek outliers menggunakan metode IQR (Interquartile Range)
print("\nDeteksi Outliers:")
# Select only numeric columns
numeric_df = df.select_dtypes(include=np.number)
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))
print(outliers.sum())  # Menampilkan jumlah outliers per kolom
```

```python
# Hitung statistik deskriptif untuk kolom 'num'
mean = df['num'].mean()
median = df['num'].median()
mode = df['num'].mode()[0]  # Ambil mode pertama jika ada beberapa
std_dev = df['num'].std()
variance = df['num'].var()
skewness = df['num'].skew()
kurtosis = df['num'].kurtosis()

# Tampilkan hasilnya
print("Mean kolom 'num':", mean)
print("Median kolom 'num':", median)
print("Mode kolom 'num':", mode)
print("Standard Deviation kolom 'num':", std_dev)
print("Variance kolom 'num':", variance)
print("Skewness kolom 'num':", skewness)
print("Kurtosis kolom 'num':", kurtosis)
```

```python
# Pilih hanya kolom numerik
numeric_df = df.select_dtypes(include=np.number)

# Hitung matriks korelasi
correlation_matrix = numeric_df.corr()

# Tampilkan matriks korelasi
print(correlation_matrix)
```

```python
# Pilih hanya kolom numerik
numeric_df = df.select_dtypes(include=np.number)

# Loop melalui setiap kolom numerik dan buat histogram dan box plot
for column in numeric_df.columns:
    plt.figure(figsize=(10, 5))  # Sesuaikan ukuran gambar jika perlu

    # Buat histogram
    plt.subplot(1, 2, 1)
    sns.histplot(numeric_df[column], kde=True)  # kde=True menambahkan kurva kepadatan
    plt.title('Histogram of ' + column)
    plt.xlabel(column)
    plt.ylabel('Frequency')

    # Buat box plot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=numeric_df[column])
    plt.title('Box Plot of ' + column)
    plt.ylabel(column)

    plt.tight_layout()  # Sesuaikan tata letak untuk menghindari tumpang tindih
    plt.show()
```

```python
# Pilih hanya kolom numerik
numeric_df = df.select_dtypes(include=np.number)

# Hitung matriks korelasi
correlation_matrix = numeric_df.corr()

# Tentukan ambang batas untuk korelasi "tinggi"
high_correlation_threshold = 0.2

# Loop melalui pasangan variabel dan buat scatter plot jika korelasi di atas ambang batas
for i in range(len(numeric_df.columns)):
    for j in range(i + 1, len(numeric_df.columns)):
        col1 = numeric_df.columns[i]
        col2 = numeric_df.columns[j]
        correlation = correlation_matrix.loc[col1, col2]
        if abs(correlation) > high_correlation_threshold:
            plt.figure(figsize=(8, 6))
            sns.scatterplot(x=col1, y=col2, data=numeric_df)
            plt.title(f"Scatter Plot of {col1} vs {col2} (Correlation: {correlation:.2f})")
            plt.show()
```
